```python
from selenium import webdriver
import csv
import time 
import pickle
from selenium.common.exceptions import StaleElementReferenceException
```

```python
# open the url
output = []
my_url = 'https://whattopair.com/'
driver = webdriver.Chrome()
driver.get(my_url)
```

```python
def count_stars(element):
    count = 0.0
    #these are the tag names for the filled and unfilled stars
    one = 'fa fa-star'
    half = 'fa fa-star-half'
    zero = 'fa fa-times'
    
    stars = element.find_elements_by_xpath('i')
    for star in stars:
        value = star.get_attribute("class")
        if value == one:
            count += 1
        elif value == half:
            count += 0.5
        elif value == zero:
            count = 0
    return count
```

```python
count = 1
max_attempts=5
while count < 974:
    attempt = 1
    while True:
        try:
            row = []
            element = driver.find_element_by_id('table_1')
            elts = element.find_elements_by_xpath('tbody/tr/td')

            #grab the text or compute the number of stars for each row
            for elt in elts:
                if elt.text == '':
                    row.append(count_stars(elt))
                else:
                    row.append(elt.text)
            output.append(row)
            break
        except StaleElementReferenceException:
            time.sleep(5)
            if attempt == max_attempts:
                raise
            attempt += 1
    
    #click the next button
    element = driver.find_element_by_id('table_1_next')
    element.click()
    print("count: ", count)
    count += 1
```

```python
#close the port
driver.close()
```

```python
outf1 = 'output.pkl'
outf2 = 'output2.pkl'
```

```python
with open(outf1, 'rb') as f1:
    data1 = pickle.load(f1)
with open(outf2, 'rb') as f2:
    data2 = pickle.load(f2)
```

```python
# save these to csv
data = []
n = 4
for page in data1:
    for i in range(10):
        data.append([str(x) for x in page[i*n:i*n+n]])
for page in data2:
    for i in range(10):
        data.append([str(x) for x in page[i*n:i*n+n]])
```

```python
with open('food_pairings.csv', 'w', newline='') as out_file:
    writer = csv.writer(out_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL, lineterminator='\n')
    writer.writerow(['Beverage Category', 'Beverage', 'Food', 'Rating'])
    for row in data:
        writer.writerow(row)
```
